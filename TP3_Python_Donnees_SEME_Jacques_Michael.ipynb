## Étape 2 — Les structures de données Python

### 1. Les Listes
Une **liste** est une collection **ordonnée** et **modifiable** d'éléments.

```python
# Déclaration d'une liste de notes
notes = [12, 15, 9, 18, 7, 14]

# Accès aux éléments
print("Première note :", notes[0])   # index 0 = premier élément
print("Dernière note :", notes[-1])  # index -1 = dernier élément

# Parcours avec une boucle for
print("\nToutes les notes :")
for note in notes:
    print(" -", note)

# Ajout d'un élément
notes.append(11)
print("\nAprès ajout de 11 :", notes)

# Suppression d'un élément
notes.remove(7)
print("Après suppression de 7 :", notes)
```

### Exercice 1 — Affichage des notes

```python
# Liste de départ
notes = [12, 15, 9, 18, 7, 14]

# Afficher TOUTES les notes
print("=== Toutes les notes ===")
for note in notes:
    print(" -", note)

# Afficher uniquement les notes >= 10
print("\n=== Notes >= 10 (admis) ===")
for note in notes:
    if note >= 10:
        print(" ✓", note)
```

### 2. Les Tuples
Un **tuple** est une collection **ordonnée** mais **immuable** (non modifiable).  
On l'utilise pour des données fixes, comme des coordonnées GPS.

```python
# Déclaration d'un tuple de coordonnées GPS (Rabat, Maroc)
coordonnees = (34.02, -6.84)

# Accès aux éléments
print("Latitude  :", coordonnees[0])
print("Longitude :", coordonnees[1])

# Différence liste vs tuple
print("\nType liste :", type([1, 2, 3]))
print("Type tuple :", type((1, 2, 3)))

# Un tuple est immuable : cette ligne provoquerait une erreur
# coordonnees[0] = 35.00  # ❌ TypeError

print("\n Tuple = données fixes, non modifiables")
```

### 3. Les Dictionnaires
Un **dictionnaire** stocke des données sous forme de paires **clé : valeur**.  
Très utilisé pour représenter des objets structurés (profil étudiant, config...).

```python
# Déclaration d'un dictionnaire étudiant
etudiant = {
    "nom"   : "Ali",
    "age"   : 21,
    "notes" : [12, 15, 14]
}

# Accès par clé
print("Nom   :", etudiant["nom"])
print("Âge   :", etudiant["age"])
print("Notes :", etudiant["notes"])

# Parcours des clés et valeurs
print("\nContenu complet :")
for cle, valeur in etudiant.items():
    print(f"  {cle} → {valeur}")
```

###  Exercice 2 — Calculer la moyenne des notes de l'étudiant

```python
# Récupération des notes depuis le dictionnaire
notes_etudiant = etudiant["notes"]

# Calcul de la moyenne
moyenne = sum(notes_etudiant) / len(notes_etudiant)

# Affichage du résultat
print(f"Étudiant : {etudiant['nom']}")
print(f"Notes    : {notes_etudiant}")
print(f"Moyenne  : {moyenne:.2f}/20")

# Mention selon la moyenne
if moyenne >= 16:
    mention = "Très Bien"
elif moyenne >= 14:
    mention = "Bien"
elif moyenne >= 12:
    mention = "Assez Bien"
elif moyenne >= 10:
    mention = "Passable"
else:
    mention = "Insuffisant"

print(f"Mention  : {mention}")
```

### 4. Les Ensembles (Sets)
Un **set** est une collection **non ordonnée** et **sans doublons**.  
Très utile pour dédupliquer des données.

```python
# Déclaration d'un ensemble ("IA" apparaît deux fois)
classes = {"IA", "DM", "Python", "IA"}

# Le set supprime automatiquement les doublons
print("Classes (sans doublons) :", classes)
print("Nombre de classes uniques :", len(classes))

# Application pratique : supprimer les doublons d'une liste
liste_avec_doublons = ["IA", "DM", "Python", "IA", "DM", "Stats"]
liste_unique = list(set(liste_avec_doublons))

print("\nListe originale   :", liste_avec_doublons)
print("Liste dédupliquée :", liste_unique)
```

## Étape 3 — Logique algorithmique appliquée aux données

### Cas pratique — Résultats d'un contrôle

```python
# Dataset simulé : résultats d'un contrôle
resultats = [8, 12, 15, 9, 18, 6, 14, 11]

print("Dataset :", resultats)
print(f"Nombre total d'étudiants : {len(resultats)}")
```

#### 1. Compter le nombre d'étudiants admis (note ≥ 10)

```python
# On filtre les notes >= 10
admis = [note for note in resultats if note >= 10]

# Comptage
nombre_admis = len(admis)

print("Notes des admis :", admis)
print(f"Nombre d'admis : {nombre_admis} / {len(resultats)}")
print(f"Taux de réussite : {nombre_admis / len(resultats) * 100:.1f}%")
```

#### 2. Calculer la moyenne générale

```python
# Calcul de la moyenne générale
moyenne_generale = sum(resultats) / len(resultats)

print(f"Somme des notes    : {sum(resultats)}")
print(f"Nombre d'étudiants : {len(resultats)}")
print(f"Moyenne générale   : {moyenne_generale:.2f}/20")
```

#### 3. Trouver la note maximale et minimale

```python
# Recherche du max et du min
note_max = max(resultats)
note_min = min(resultats)

print(f"Note maximale : {note_max}/20")
print(f"Note minimale : {note_min}/20")
print(f"Écart         : {note_max - note_min} points")
```

#### 4. Identifier les notes inférieures à la moyenne

```python
# Filtrer les notes sous la moyenne générale
sous_moyenne = [note for note in resultats if note < moyenne_generale]

print(f"Moyenne générale       : {moyenne_generale:.2f}")
print(f"Notes sous la moyenne  : {sous_moyenne}")
print(f"Nombre d'étudiants concernés : {len(sous_moyenne)}")
```

#### 5. Bilan complet du contrôle

```python
# Récapitulatif de tous les résultats
print("=" * 40)
print("        BILAN DU CONTRÔLE")
print("=" * 40)
print(f"  Dataset          : {resultats}")
print(f"  Nb étudiants     : {len(resultats)}")
print(f"  Moyenne générale : {moyenne_generale:.2f}/20")
print(f"  Note max         : {note_max}/20")
print(f"  Note min         : {note_min}/20")
print(f"  Admis (>= 10)    : {nombre_admis} étudiants")
print(f"  Taux de réussite : {nombre_admis / len(resultats) * 100:.1f}%")
print(f"  Sous la moyenne  : {sous_moyenne}")
print("=" * 40)
```
